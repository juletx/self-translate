# Translation metrics

```python
import matplotlib.pyplot as plt

def plot_size_df(df, model_sizes, model_names, title):
    dft = df.set_index("lang").T
    dft["size"] = model_sizes
    dft.plot(x="size", marker="o")
    plt.title(title)
    plt.xticks(model_sizes, model_sizes, rotation="vertical");
```

## Zero-shot translation

```python
import pandas as pd

metrics_df = pd.read_csv("translation_metrics_xstory_cloze_mt.tsv", sep="\t")
```

```python
metrics_df
```

### BLEU

```python
# plot nllb models
metrics_df[metrics_df["model"].str.startswith("nllb")].pivot(index="lang", columns="model", values="bleu").plot.bar()
```

```python
# plot bloomz models
metrics_df[metrics_df["model"].str.startswith("bloomz")].pivot(index="lang", columns="model", values="bleu").plot.bar()
```

```python
# plot bloom models
metrics_df[metrics_df["model"].str.startswith("bloom-")].pivot(index="lang", columns="model", values="bleu").plot.bar()
```

```python
# plot xglm models
metrics_df[metrics_df["model"].str.startswith("xglm-")].pivot(index="lang", columns="model", values="bleu").plot.bar()
```
